```python
distinct_location = sqldf('''
select distinct location from df_covid_all where location = "Bahrain"
''')
distinct_location
```

```python
df_bahrain = sqldf('''
select * from df_covid_all where location = "Bahrain" and total_cases IS NOT NULL and total_deaths IS NOT NULL
''')
df_bahrain
```

```python
count_data = df_bahrain.shape[0]
count_data
```
